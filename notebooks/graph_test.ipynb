```python
# Example usage function
def run_identity_search(airbnb_url: str, query: str = "") -> dict:
    """
    Run the multi-agent identity search system.
    
    Args:
        airbnb_url: URL to Airbnb profile or listing
        query: Additional search context or specific requirements
    
    Returns:
        Dictionary containing the final report and analysis
    """
    graph = create_multi_agent_graph()
    
    initial_state = {
        "airbnb_url": airbnb_url,
        "query": query,
        "airbnb_profile_url": None,
        "airbnb_host_info": None,
        "airbnb_listings": None,
        "airbnb_reviews": None,
        "search_keywords": [],
        "location_keywords": [],
        "instagram_profiles": [],
        "final_report": {},
        "current_step": "starting",
        "errors": []
    }
    
    result = graph.invoke(initial_state)
    return result

```

```python

# Example usage
result = run_identity_search(
    airbnb_url="https://www.airbnb.com/users/show/395895315",
    query="Looking for Instagram profiles of this Airbnb host in Morocco"
)
```

```python

if result.get("final_report"):
    print("=== IDENTITY SEARCH REPORT ===")
    print(result["final_report"]["summary"])
    print(f"\nTop Matches Found: {len(result['final_report']['top_matches'])}")
    
    for i, match in enumerate(result["final_report"]["top_matches"][:3], 1):
        print(f"\n{i}. @{match['username']}")
        print(f"   Name: {match['full_name']}")
        print(f"   Confidence: {match['confidence_score']:.1%}")
        print(f"   Followers: {match['followers']:,}")
        print(f"   Verified: {match['is_verified']}")
        print(f"   Reasons: {', '.join(match['match_reasons'][:3])}")
else:
    print("Search completed with errors:")
    for error in result.get("errors", []):
        print(f"- {error}")
```
